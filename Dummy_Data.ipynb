#1. Solution for the Case study


> Generated the dummy data


> Make it 500 samples of data






```python
import numpy as np
import pandas as pd
from datetime import timedelta, datetime

# Set a seed for reproducibility
np.random.seed(42)

# Number of data points
num_points = 500

# Generate a date range from January 1, 2023, to April 10, 2024
start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 4, 10)
date_range = [start_date + timedelta(days=i) for i in range(num_points)]

# Create dummy data for each variable
defect_density = np.random.uniform(low=0.1, high=5, size=num_points)
downtime_minutes = np.random.uniform(low=0, high=120, size=num_points)
scrap_rate = np.random.uniform(low=0, high=10, size=num_points)
customer_complaint_rate = np.random.uniform(low=0, high=5, size=num_points)
cpk = np.random.uniform(low=0.5, high=2, size=num_points)
rework_rate = np.random.uniform(low=0, high=8, size=num_points)

# Generate dummy data for categories and quality issues
categories = ["Welding", "Material Quality", "Surface Finish", "Dimensional Accuracy",
              "Structural Integrity", "Assembly Process", "Non-Destructive Testing (NDT)"]

quality_issues = ["Welding Defects", "Material Issues", "Surface Imperfections", "Dimensional Deviations",
                  "Structural Weakness", "Coating and Painting Problems", "Assembly Issues",
                  "Non-Conformities in NDT"]

# Assign random quality issues to each category for each date
category_data = np.random.choice(categories, size=num_points)
issue_data = np.random.choice(quality_issues, size=num_points)

# Create a DataFrame for the categories and quality issues
category_quality_data = pd.DataFrame({
    'Date': date_range,
    'Category': category_data,
    'Quality_Issue': issue_data
})

# Combine with the previously generated data
combined_data = pd.DataFrame({
    'Date': date_range,
    'Defect_Density': defect_density,
    'Downtime_Minutes': downtime_minutes,
    'Scrap_Rate': scrap_rate,
    'Customer_Complaint_Rate': customer_complaint_rate,
    'Process_Capability_Index': cpk,
    'Rework_Rate': rework_rate
})

# Merge the data on the 'Date' column
final_data = pd.merge(combined_data, category_quality_data, on='Date')

# Export the data to an Excel file
#final_data.to_excel('Latest dummy data.xlsx', index=False)

# Display the first few rows of the combined data
print(final_data.head(10))
```

#2. Recommendation - Vendor Selection

```python
import pandas as pd

# Number of vendors
num_vendors = 50

# Generate vendor names as "Vendor 1," "Vendor 2," ...
vendor_names = [f"Vendor {i}" for i in range(1, num_vendors + 1)]

# Quality issues categories based on monopile components
quality_issues = np.random.choice(['Material Issues', 'Surface Imperfections', 'Non-Conformities in NDT',
                                   'Coating and Painting Problems', 'Welding Defects', 'Structural Weakness',
                                   'Dimensional Deviations'], size=num_vendors)

# Generate random downtime minutes for each vendor
downtime_minutes = np.random.uniform(low=0, high=300, size=num_vendors)

# Generate random defect types for each vendor
defect_types = np.random.choice(['Rejected', 'No Impact', 'Impact'], size=num_vendors)

# Generate random material types for each vendor
material_types = np.random.choice(['Raw Materials', 'Cables', 'Components', 'Metal Sheets', 'Electronics'], size=num_vendors)

# Additional metrics
defect_density = np.random.uniform(low=0.1, high=5, size=num_vendors)
cpk = np.random.uniform(low=0.5, high=2, size=num_vendors)
rework_rate = np.random.uniform(low=0, high=8, size=num_vendors)

# Generate random scrap rates for each vendor
scrap_rate = np.random.uniform(low=0, high=10, size=num_vendors)

# Create a DataFrame to organize the data
data = pd.DataFrame({
    'Vendor_Name': vendor_names,
    'Quality_Issues': quality_issues,
    'Downtime_Minutes': downtime_minutes,
    'Defect_Type': defect_types,
    'Material_Type': material_types,
    'Defect_Density': defect_density,
    'Process_Capability_Index': cpk,
    'Rework_Rate': rework_rate,
    'Scrap_Rate': scrap_rate
})

# Display the generated data
print(data.head(10))

# Export the data to an Excel file
#data.to_excel('Vendor dummy data.xlsx', index=False)

# Display the first few rows of the combined data
print(data.head(10))
```
